```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
df = df.drop(columns=['Time'])
df.head()
```

```python
X=df.drop(columns=['Class'])
y=df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

1. **Undersampling and Oversampling:**

---


   Since the dataset is highly imbalanced (0.173% minority class), only Undersampling would lead to huge data loss and only Oversampling will increase the computational complexity and overfit.

   So, we first undersample the data to 10000 entries and then oversample the minority to 10000

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
rus = RandomUnderSampler(sampling_strategy={0:10000,1:394}, random_state=42)
X_rus,y_rus=rus.fit_resample(X_train,y_train)
ros=RandomOverSampler(sampling_strategy='minority',random_state=42)
X_rus_ros,y_rus_ros=ros.fit_resample(X_rus,y_rus)
```

```python
sns.countplot(x=y_rus_ros)
plt.title("Distribution of Target Classes")
plt.xlabel("Target Class")
plt.ylabel("Count")
plt.show()
```

2. SMOTE **Analysis**

---



```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_smote,y_smote=smote.fit_resample(X_train,y_train)
y_smote.value_counts()
```

```python
sns.countplot(x=y_smote)
plt.title("Distribution of Target Classes")
plt.xlabel("Target Class")
plt.ylabel("Count")
plt.show()
```

3. **Balanced RF:**

---

In this case while performing bagging of the data we keep the minority class same and only sample the majority class for each estimator

```python
from imblearn.ensemble import BalancedRandomForestClassifier

brf = BalancedRandomForestClassifier(n_estimators=100, random_state=42)
```

**4. Random Forest Classifier:**

---
Here we just balance the class weights


```python
from sklearn.ensemble import RandomForestClassifier

rf_clf= RandomForestClassifier(n_estimators=100,class_weight='balanced')
```

```python
models={
    'UnderSampling and OverSampling':RandomForestClassifier(n_estimators=100,random_state=42).fit(X_rus_ros,y_rus_ros),
    'SMOTE':RandomForestClassifier(n_estimators=100,random_state=42).fit(X_smote,y_smote),
    'Balanced RF':brf.fit(X_train,y_train),
    'Cost-Sensitive':rf_clf.fit(X_train,y_train),
    'Baseline (No Sampling)':RandomForestClassifier(n_estimators=100,random_state=42).fit(X_train,y_train)
}
```

```python
from sklearn.metrics import classification_report,roc_auc_score,average_precision_score
results ={}

for name,model in models.items():
  y_pred=model.predict(X_test)
  y_prob=model.predict_proba(X_test)[:,1]
  results[name] = {'Classification Report':classification_report(y_test,y_pred,output_dict=True),
                   'ROC_AUC':roc_auc_score(y_test,y_prob),
                   'PR_AUC':average_precision_score(y_test,y_prob)
                   }
```

```python
results_df =pd.DataFrame({
    'Method':[],
    'Minority Recall':[],
    'Minority Precision':[],
    'Minority F1':[],
    'ROC_AUC':[],
    'PR_AUC':[]
})

for name,metrics in results.items():
  new_row_df=pd.DataFrame([{
      'Method':name,
      'Minority Recall':metrics['Classification Report']['1']['recall'],
      'Minority Precision':metrics['Classification Report']['1']['precision'],
      'Minority F1':metrics['Classification Report']['1']['f1-score'],
      'ROC_AUC':metrics['ROC_AUC'],
      'PR_AUC':metrics['PR_AUC']
  }])
  results_df = pd.concat([results_df, new_row_df], ignore_index=True)
results_df
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
for name,model in models.items():
  y_prob=model.predict_proba(X_test)[:,1]
  fpr,tpr,_=roc_curve(y_test,y_prob)
  plt.plot(fpr,tpr,label=f'{name} (AUC = {roc_auc_score(y_test,y_prob):.2f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
for name,model in models.items():
  y_pred=model.predict(X_test)
  cm=confusion_matrix(y_test,y_pred)
  sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
  plt.title(f'Confusion Matrix:{name}')
  plt.show()
```
